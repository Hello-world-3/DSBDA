## Data integration

```python
data1=["Country","City","AQI Value","AQI Category","CO AQI Value"]
data2=["Ozone AQI Value","Ozone AQI Category","NO2 AQI Value","NO2 AQI Category","PM2.5 AQI Value","PM2.5 AQI Category"]
```

```python
part1=df[data1]
part2=df[data2]
```

```python
part1
```

```python
part2
```

```python
merge_df=pd.concat([part1,part2],axis=1)
```

```python
merge_df
```

```python
# ERROR CORRECTION

# Calculate the 99th percentile of AQI values
import numpy as np
percentile_99 = np.percentile(merge_df['AQI Value'], 99)

# Define the threshold for outliers
threshold = percentile_99

# Calculate the mean and standard deviation of AQI values
mean_aqi = merge_df['AQI Value'].mean()
std_aqi = merge_df['AQI Value'].std()

# Identify outliers using the threshold
lower_bound = mean_aqi - threshold * std_aqi
upper_bound = mean_aqi + threshold * std_aqi

# Remove outliers from the dataset
merge_df = merge_df[(merge_df['AQI Value'] >= lower_bound) & (merge_df['AQI Value'] <= upper_bound)]

print("Outliers removed.")
```
